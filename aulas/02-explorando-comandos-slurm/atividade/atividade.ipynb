```python
from matplotlib import pyplot as plt
```

```python
n_its = [0, 15, 30, 50, 100]
t_limitado = [0, 0.86, 1.73, 2.86, 5.74]
t_buffado = [0, 0.23, 0.46, 0.75, 1.51]

n_tasks = list(range(1, 7))
n_cpus = list(range(1, 7))
memory = [512, 1024, 2048]
tasks_vs_t_custom = [6.29, 2.93, 2.02, 1.51, 1.28, 1.09]
cpus_vs_t_custom = [6.29, 5.61, 5.73, 5.87, 5.81, 5.84]
memory_vs_t_custom = [6.29, 5.75, 9.95]
```

```python
plt.plot(n_its, t_limitado)
plt.plot(n_its, t_buffado)
plt.legend(["Limitada", "Melhorada"])
plt.title("Tempo de execu√ß√£o em fun√ß√£o do n√∫mero de\nitera√ß√µes para diferentes configura√ß√µes do job")
plt.xlabel("Itera√ß√µes")
plt.ylabel("Tempo (s)")
plt.show()
```

```python
plt.plot(n_tasks, tasks_vs_t_custom)
plt.plot(n_cpus, cpus_vs_t_custom)
plt.legend(["Tasks", "CPUs"])
plt.title("Tempo de execu√ß√£o em fun√ß√£o da quantidade de tasks e de CPUs")
plt.xlabel("Quantidade")
plt.ylabel("Tempo (s)")
plt.show()
```

```python
plt.plot(memory, memory_vs_t_custom, c="g")
plt.title("Tempo de execu√ß√£o em fun√ß√£o da quantidade de mem√≥ria")
plt.xlabel("Mem√≥ria (kB)")
plt.ylabel("Tempo (s)")
plt.show()
```
