Otimização e Profiling

```python
from matplotlib import pyplot as plt
```

```python
n_its = [0, 30, 50, 100]
t_O0 = [0, 1.70, 2.81, 5.62]
t_O1 = [0, 0.46, 0.77, 1.54]
t_O2 = [0, 0.36, 0.60, 1.19]
t_O3 = [0, 0.16, 0.27, 0.53]
```

```python
plt.plot(n_its, t_O0, label="O0")
plt.plot(n_its, t_O1, label="O1")
plt.plot(n_its, t_O2, label="O2")
plt.plot(n_its, t_O3, label="O3")
plt.legend()
plt.title("Tempo de execução para diferentes níveis de otimização no compilador")
plt.xlabel("Iterações")
plt.ylabel("Tempo (s)")
plt.show()
```
