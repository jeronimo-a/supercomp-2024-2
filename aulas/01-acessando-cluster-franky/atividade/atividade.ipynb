**Resultados PC**

**Resultados Cluster Franky**

| Tamanho | 300x300 | 900x900 | 1300x1300 |

| C++ paralelo | 0,04 s | 1,15 s | 3,49 s |

**Gráfico**

```python
from matplotlib import pyplot as plt
```

```python
sizes = [300, 900, 1300]
pc_python = [2.79, 76.84, 229.03]
pc_cpp_seq = [0.19, 5.09, 15.87]
pc_cpp_par = [0.04, 0.81, 2.42]
cluster_python = [2.41, 68.93, 226.73]
cluster_cpp_seq = [0.14, 4.02, 12.53]
cluster_cpp_par = [0.04, 1.15, 3.49]
```

```python
plt.plot(sizes, pc_python)
plt.plot(sizes, pc_cpp_seq)
plt.plot(sizes, pc_cpp_par)
plt.legend(["Python", "C++ sequencial", "C++ paralelo"])
plt.title("Tempo de execução em função do tamanho da matriz no PC")
plt.ylabel("Tempo (s)")
plt.xlabel("Tamanho da matriz quadrada")
plt.show()
```

```python
plt.plot(sizes, cluster_python)
plt.plot(sizes, cluster_cpp_seq)
plt.plot(sizes, cluster_cpp_par)
plt.legend(["Python", "C++ sequencial", "C++ paralelo"])
plt.title("Tempo de execução em função do tamanho da matriz no cluster Franky")
plt.ylabel("Tempo (s)")
plt.xlabel("Tamanho da matriz quadrada")
plt.show()
```

```python
plt.plot(sizes, pc_python)
plt.plot(sizes, cluster_python)
plt.legend(["PC", "Cluster"])
plt.title("Tempo de execução Python em função do tamanho da matriz")
plt.ylabel("Tempo (s)")
plt.xlabel("Tamanho da matriz quadrada")
plt.show()
```

```python
plt.plot(sizes, pc_cpp_seq)
plt.plot(sizes, cluster_cpp_seq)
plt.legend(["PC", "Cluster"])
plt.title("Tempo de execução C++ sequencial em função do tamanho da matriz")
plt.ylabel("Tempo (s)")
plt.xlabel("Tamanho da matriz quadrada")
plt.show()
```

```python
plt.plot(sizes, pc_cpp_par)
plt.plot(sizes, cluster_cpp_par)
plt.legend(["PC", "Cluster"])
plt.title("Tempo de execução C++ paralelizado em função do tamanho da matriz")
plt.ylabel("Tempo (s)")
plt.xlabel("Tamanho da matriz quadrada")
plt.show()
```

```python
pc_max_diff = list()
cluster_max_diff = list()

for i in range(len(pc_python)):
    print(pc_cpp_par[i], pc_python[i])
    pc_max_diff.append(pc_cpp_par[i] / pc_python[i] * 100 - 1)
    cluster_max_diff.append(cluster_cpp_par[i] / cluster_python[i] * 100 - 1)
```

```python
plt.plot(sizes, pc_max_diff)
plt.plot(sizes, cluster_max_diff)
plt.legend(["PC", "Cluster"])
plt.title("")
plt.ylabel("Tempo (s)")
plt.xlabel("Tamanho da matriz quadrada")
plt.show()
```

**Comparação**

A partir dos resultados conclui-se que, neste contexto, a influência da linguagem de programação na performance do programa é maior que e da paralelização, visto que a diferença de tempo de execução entre o programa em C++ paralelizado e sequencial é menor que a diferença entre o programa em Python e o sequencial em C++.

Conforme a entrada aumente, todas as diferenças de tempo de execução seriam cada vez maiores.

Em um ambiente de HPC, este problema pode ser abordado sempre utilizando a linguagem mais adequada e paralelizando o algoritmo ao máximo.
